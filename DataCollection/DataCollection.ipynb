```python
# maybe change some of these to reduce redundancy
basic_cols = ['Age', 'Tm', 'G', 'GS', 'MP', 'FGA',
       '2PA', '3PA', 'FTA', 'ORB', 'DRB', 'AST',
       'STL', 'BLK', 'TOV', 'PF', 'PTS', 'FG%', '2P%', '3P%', 'eFG%', 'FT%',
       'TS%']
```

```python
adv_cols = [ 'PER', '3PAr',
       'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%',
       'ORtg', 'DRtg', 'OWS', 'DWS', 'WS', 'WS/48', 'OBPM', 'DBPM', 'BPM',
       'VORP']
```

```python
advdf = stats_adv.copy()
```

```python
for col in basic_cols:
    df[col + '_year1'] = 0
    df[col + '_year2'] = 0
    
for col in adv_cols:
    df[col + '_year1'] = 0
    df[col + '_year2'] = 0
```

```python
df.columns
```

```python
df = df.sort_values(by=['Season'])
advdf = advdf.sort_values(by=['Season'])
```

```python
df.head()
```

```python
for player in df['Player'].unique():
    playerstats = df.loc[df['Player'] == player].reset_index(drop = True)
    advplayerstats = advdf.loc[advdf['Player'] == player].reset_index(drop = True)
    if len(playerstats) < 2 or len(advplayerstats) < 2:
        df = df.drop(df[df['Player'] == player].index).reset_index(drop = True)
        advdf = advdf.drop(advdf[advdf['Player'] == player].index).reset_index(drop = True)
        continue
    else:
        for col in basic_cols:
            df[col + "_year1"][df[df['Player'] == player].index] = playerstats[col][0]
            df[col + "_year2"][df[df['Player'] == player].index] = playerstats[col][1]
        for col in adv_cols:
            df[col + "_year1"][df[df['Player'] == player].index] = advplayerstats[col][0]
            df[col + "_year2"][df[df['Player'] == player].index] = advplayerstats[col][1]
```

```python
df = df.sort_values(by=['Season']).drop_duplicates(subset ="Player",keep='first').reset_index(drop=True)
```

```python
df.drop(columns=[*basic_cols, 'Lg', 'FG', '2P', '3P', 'FT', 'TRB'], inplace = True)
```

```python
rising_stars = pd.DataFrame()
```

```python
for year in range(2000,2021):
    # URL page we will scraping (see image above)
    url = "https://www.basketball-reference.com/allstar/NBA_{}_rookie.html".format(year)
    # this is the HTML from the given URL
    html = urlopen(url)
    soup = BeautifulSoup(html)
    # use findALL() to get the column headers
    soup.findAll('tr', limit=2)
    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in soup.findAll('tr', limit=6)[5].findAll('th')]
    # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers
    # avoid the first header row
    rows = soup.findAll('tr')
    players = [[td.getText() for td in rows[i].findAll(['td','th'])] for i in range(len(rows))]
    rising_stars = pd.concat([rising_stars, pd.DataFrame(players, columns = headers)]).reset_index(drop=True)
```

```python
risingstarsDict = {}
for player in rising_stars['Starters']:
    risingstarsDict[player] = 1
```

```python
df['rising_star'] = df['Player'].apply(lambda x: 1 if x in risingstarsDict else 0)
```
