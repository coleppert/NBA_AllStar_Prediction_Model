```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
stats = pd.DataFrame()
```

```python
for offset in range(0,32):
    # URL page we will scraping (see image above)
    url = "https://www.basketball-reference.com/play-index/psl_finder.cgi?request=1&match=single&type=per_game&per_minute_base=36&per_poss_base=100&lg_id=NBA&is_playoffs=N&year_min=1998&year_max=2020&franch_id=&season_start=1&season_end=2&age_min=0&age_max=99&shoot_hand=&height_min=0&height_max=99&birth_country_is=Y&birth_country=&birth_state=&college_id=&draft_year=&is_active=&debut_yr_nba_start=&debut_yr_nba_end=&is_hof=&is_as=&as_comp=gt&as_val=0&award=&pos_is_g=Y&pos_is_gf=Y&pos_is_f=Y&pos_is_fg=Y&pos_is_fc=Y&pos_is_c=Y&pos_is_cf=Y&qual=&c1stat=&c1comp=&c1val=&c2stat=&c2comp=&c2val=&c3stat=&c3comp=&c3val=&c4stat=&c4comp=&c4val=&c5stat=&c5comp=&c6mult=&c6stat=&order_by=player&order_by_asc=&offset={}".format(str(offset*100))
    # this is the HTML from the given URL
    html = urlopen(url)
    soup = BeautifulSoup(html)
    # use findALL() to get the column headers
    soup.findAll('tr', limit=2)
    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in soup.findAll('tr', limit=2)[1].findAll('th')]
    # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers[1:]
    # avoid the first header row
    rows = soup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
                for i in range(len(rows))]
    
    stats = pd.concat([stats, pd.DataFrame(player_stats, columns = headers)]).reset_index(drop=True)
```

```python
df = stats.copy()
df.columns
```

```python
cols_to_average = [ 'Age', 'G', 'GS', 'MP', 'FG', 'FGA',
       '2P', '2PA', '3P', '3PA', 'FT', 'FTA', 'ORB', 'DRB', 'TRB', 'AST',
       'STL', 'BLK', 'TOV', 'PF', 'PTS']
```

```python
for col in cols_to_average:
    df[col + "_diff"] = 0
    df[col + "_avg"] = 0
```

```python
for player in df['Player'].unique():
    pstats = df.loc[df['Player'] == player].reset_index(drop = True)
    if len(pstats) < 2:
        df = df.drop(df[df['Player'] == player].index).reset_index(drop = True)
        continue
    else:
       
        for col in cols_to_average:
            #print(col)
            df[col + "_diff"][df[df['Player'] == player].index] = float(pstats[col][1]) - float(pstats[col][0])
            df[col + "_avg"][df[df['Player'] == player].index] = (float(pstats[col][1]) + float(pstats[col][0])) / 2
```

```python
df.head()
```

```python
df = df.sort_values(by=['Season']).drop_duplicates(subset ="Player",keep='first').reset_index(drop=True)
```
